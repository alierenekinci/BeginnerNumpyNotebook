```python
import numpy as np  # numpy modulunun np lakabiyla import edilmesi
a = np.linspace(1,10,10) #ilerleyen kısımlarda tekrar anlatacağım
# linspace reel sayilardan (float) mutesekkil bir dizi yaratir
print("a: ", a, end="\n\n")
print("a[:5]: ", a[:5], end="\n\n")
print("a[7:]: ", a[7:], end="\n\n")
print("a[3:7]: ", a[3:7], end="\n\n")
print("a[0:-1:2]: ", a[0:-1:2], end="\n\n")
print("a[::4]: ", a[::4], end="\n\n")
print("a[1:-1]: ", a[1:-1], end="\n\n")
#bölümlediğimiz a matrisini b ye eşitledik
b = a[1:-1]
print("b: ", b, end="\n\n")
#Dikkat burda benin ilinci elemanını 0.1 yaptık bak konrol et hadi
b[2] = 0.1
print("Yeni b: ", b, end="\n\n")
print("a: ", a, end="\n\n")
```

## zeros
(elemanların tamamı 0 olan dizi oluşturur)

```python
d1 = np.zeros(5)
#içerisine shape alır. Shapeler ise bir tupledır.
# birden çok argüman gireceğimiz zaman tuple kullanmamız gerekli
d1
```

```python
print(np.zeros((5,)))#tuple yarattık (5,)
```

```python
print(np.zeros((5,2))) #tuple yarattık (5,2) iki boyutlu 0 matris oluşturduk.
```

```python
print(np.zeros((5,2,3,4,219,)))#tuple yarattık (5,2,3) üç boyutlu 0 matris oluşturduk.
```

```python
# d40 benim bilgisayarımda şu şekildeydi
# array([[4.9e-324, 9.9e-324, 1.5e-323, 2.0e-323]])
d40.fill(3.3)
d40
#artık
#array([[3.3, 3.3, 3.3, 3.3]])
```
